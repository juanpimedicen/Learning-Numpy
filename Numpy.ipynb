n = np.zise(Moscu)

```python
n = np.size(Moscu)
y = Moscu
x = fechas
```

```python
a = (n*np.sum(x*y) - np.sum(x)*np.sum(y))/(n*np.sum(x**2)-np.sum(x)**2)
```

```python
b = np.mean(y) - a*np.mean(x)
```

```python
print(f'La pendiente es: {a}')
print(f'El intercepto es: {b}')
```

```python
y = a*x + b
y
```

```python
np.linalg.norm(Moscu - y)
```

```python
plt.plot(fechas,Moscu)
plt.plot(x,y)
plt.plot(25,a*25+b,'ro')
plt.plot(75,a*75+b,'*b')
```
